```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
```

```python
import matplotlib.pyplot as plt

def plot_tree(root, x=0, y=0, parent_x=None, parent_y=None, level=0, spacing=1.5):
    if not root:
        return

    if parent_x is not None and parent_y is not None:
        plt.plot([parent_x, x], [parent_y, y], '-k', lw=2)

    plt.plot(x, y, 'ok', ms=15)
    plt.text(x, y, str(root.val), va='center', ha='center', color='white', fontsize=12)

    left_spacing = right_spacing = spacing

    if root.left:
        left_spacing = spacing / 2
        plot_tree(root.left, x=x - left_spacing, y=y - 1, parent_x=x, parent_y=y, level=level + 1, spacing=left_spacing)

    if root.right:
        right_spacing = spacing / 2
        plot_tree(root.right, x=x + right_spacing, y=y - 1, parent_x=x, parent_y=y, level=level + 1, spacing=right_spacing)

    if level == 0:
        plt.axis('off')
        plt.show()
```

```python
root = Node(7)
root.left = Node(3)
root.right = Node(12)
root.left.left = Node(99)
root.left.right = Node(8)
root.right.left = Node(5)
root.right.right = Node(6)
```

```python
plt.figure(figsize=(8,8))
plot_tree(root)
plt.axis('off')
plt.show()
```

```python
def printMaxLevel(root):
    if not root:
        return
    if not root.left and not root.right:
        print(root.val)
    elif not root.right or root.left.val > root.right.val:
        print(root.val)
        printMaxLevel(root.left)
    else:
        print(root.val)
        printMaxLevel(root.right)
```
